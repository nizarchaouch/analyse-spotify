### Quels ont t les genres les plus importants de la dcennie ?

```python
df_top=df.groupby('top genre')['top genre'].count().sort_values(ascending=False)
df_top.head(10)
```

Le tableau ci-dessus nous indique que le genre dance pop a t le genre le plus populaire tout au long de la dcennie.

```python
genres_year = df[['top genre', 'year']]
def cont_gen(ds):
    years = range(2010, 2020)
    y_g = pd.DataFrame()
    for i in years:
        col = ds[ds.year == i]['top genre'].value_counts()
        col = col[:3]
        col.to_frame()
        col = col.rename(str(i))
        y_g = pd.concat([y_g,col]) 

    return y_g

genres_years = cont_gen(genres_year)
genres_years
```

```python
df['top genre'].value_counts().head(10).plot.pie(figsize=(7,7), autopct='%1.0f%%')
```

```python
f1 = plt.figure()
colors = ['tomato','lightseagreen', 'Aqua', 'Firebrick', 'DarkViolet', 'Gainsboro', 'Olive', 'Gold', 'Pink', 'Black', 'Blue', 'Red']
plt.style.use('seaborn-darkgrid')
genres_years.plot(kind='bar', ax=f1.gca(), stacked=True ,figsize=(7,5), color=colors, ylabel='Nombre de chansons dans le top', xlabel='Annes', title='Les genres les plus populaires de la dcennie')
plt.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))
plt.show()
```

Il est incontestable que la dance pop a domin tout au long de la dcennie. En plus de la "dance pop", nous pouvons voir que des genres tels que la "pop" et la "pop canadienne" ont galement t pertinents au cours de la dcennie. Il ne fait aucun doute que, globalement, le genre pop et toutes ses variantes ont domin la dcennie.

## Les chansons sont-elles devenues plus tristes ?

Pour obtenir le changement d'humeur des chansons tout au long de l'anne, ce qui a t fait tait d'obtenir la moyenne de la variable val dans une anne, puis de tracer la valeur de chaque anne respective.

```python
sad = df[['val', 'year']]
def cont_sad(ds):
    years = range(2010, 2020)
    s = {}
    for i in years:
        sa = sad[sad.year == i]
        pro = sa.mean()
        s[i] = pro[0]
    return s
sad1 = cont_sad(sad)
sad1 = pd.DataFrame(sad1.items(), columns=['year', 'prom_sad'])
sad1
```

```python
f3 = sns.lineplot(data=sad1, x="year", y="prom_sad")
f3.set_title("Graphique de l'volution de l'ambiance des chansons dans la dcennie 2010-2019")
f3.set_ylabel('Positivit des chansons')
f3.set_xlabel('Annes')
plt.show()
```

Au milieu de la dcennie, il y a eu une nette diminution de l'ambiance des chansons et cela s'est poursuivi jusqu' la fin de la dcennie, tant l'anne 2017, l'anne o les chansons redeviendraient positives.

## Comment la variable nergie a-t-elle volu ?

```python
nrgy = df[['nrgy', 'year']]
def cont_nrgy(ds):
    years = range(2010, 2020)
    s = {}
    for i in years:
        sa = nrgy[nrgy.year == i]
        pro = sa.mean()
        s[i] = pro[0]
    return s
energy = cont_nrgy(nrgy)
energy = pd.DataFrame(energy.items(), columns=['year', 'prom_energy'])
energy
```

```python
f4 = sns.lineplot(data=energy, x="year", y="prom_energy")
f4.set_title("Graphique du changement d'nergie des chansons dans la dcennie 2010-2019")
f4.set_ylabel("L'nergie des chansons")
f4.set_xlabel('Annes')
plt.show()
```

Comme pour la variable d'humeur, l'nergie des chansons a galement diminu au cours de la dcennie.

## Comment la variable dansabilit a-t-elle volu ?

```python
dnce = df[['dnce', 'year']]
def cont_dnce(ds):
    years = range(2010, 2020)
    s = {}
    for i in years:
        sa = dnce[dnce.year == i]
        pro = sa.mean()
        s[i] = pro[0]
    return s
dance = cont_dnce(dnce)
dance = pd.DataFrame(dance.items(), columns=['year', 'prom_dance'])
dance
```

```python
f5 = sns.lineplot(data=dance, x="year", y="prom_dance")
f5.set_title("Graphique de l'volution de la dansabilit des chansons dans la dcennie 2010-2019")
f5.set_ylabel('Dansabilit des chansons')
f5.set_xlabel('Annes')
plt.show()
```

Curieusement, bien que le changement d'nergie et d'humeur des chansons ait t ngatif, la danse des chansons montre une augmentation  la fin de la dcennie.
